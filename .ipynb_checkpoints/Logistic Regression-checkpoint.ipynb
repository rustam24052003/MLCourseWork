# **Import libraries**


```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt # data visualization
import seaborn as sns # statistical data visualization
%matplotlib inline
from sklearn.linear_model import LogisticRegression
from  sklearn.naive_bayes  import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
import category_encoders as ce
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report
```

# ** Import dataset** <a class="anchor" id="8"></a>



```python
data = '/kaggle/input/car-evaluation-data-set/car_evaluation.csv'
#data = '/content/drive/MyDrive/car_evaluation.csv'
df = pd.read_csv(data, header=None)
df
```

```python
# view dimensions of dataset

df.shape
```

### Rename column names

We can see that the dataset does not have proper column names. The columns are merely labelled as 0,1,2.... and so on. We should give proper names to the columns. I will do it as follows:-

```python
col_names = ['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'class']


df.columns = col_names

col_names
```

```python
# let's again preview the dataset

df.head()
```

```python
X = df.drop(['class'], axis=1)
y = df['class']
y
```

# **Split data into separate training and test set** <a class="anchor" id="11"></a>


```python
# split X and y into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```python
# check the shape of X_train and X_test

X_train.shape, X_test.shape
```

```python
X_train.head()
```

```python
# encode variables with ordinal encoding
encoder = ce.OrdinalEncoder(cols=['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety'])
X_train = encoder.fit_transform(X_train)
X_test = encoder.transform(X_test)
```

```python
X_train.head()
```

```python
X_train.shape
```

```python
X_test.head()
```

```python
X_test.shape
```

```python

clf_gini = LogisticRegression()
# fit the model
clf_gini.fit(X_train, y_train)
```

### Predict the Test set results with criterion gini index

```python
y_pred_gini = clf_gini.predict(X_test)
y_pred_gini
```

### Check accuracy score

```python
print('Model accuracy score: {0:0.4f}'. format(accuracy_score(y_test, y_pred_gini)))
```

# ** Confusion matrix** <a class="anchor" id="15"></a>




```python
cm = confusion_matrix(y_test, y_pred_gini)
plt.figure(figsize=(4,4))
sns.heatmap(cm,annot=True,fmt='0.1f', linewidth = 0.1 , square = True , cbar = False)
plt.ylabel("Prediction Value")
plt.xlabel("Actual Value")
plt
```

```python
print( classification_report(y_test, y_pred_gini))
```
